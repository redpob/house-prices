```python
categorical_columns = train.drop(train.describe().columns, axis = 1)
for col in categorical_columns.columns: 
    fig = plt.figure(figsize = (10, 6))
    sns.boxplot(x = col, y = 'SalePrice', data = train)
```

```python
dummy_all = pd.DataFrame()
for col in categorical_columns.columns: 
    dummy = pd.get_dummies(train[col], prefix = col)
    if dummy_all.empty: 
        dummy_all = dummy
    else: 
        dummy_all = dummy_all.merge(dummy, how = 'outer', left_index = True, right_index = True)
new_train = dummy_all.merge(train[numerical_cols], how = 'outer', left_index = True, right_index = True)
x = new_train.drop('SalePrice', axis = 1).drop([
    'Id', 'MSSubClass', 'OverallCond', 'BsmtFinSF2', 'LowQualFinSF', 'BsmtHalfBath', 
    'KitchenAbvGr', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 
    'MoSold', 'YrSold'], axis = 1)
y = new_train['SalePrice']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn import metrics
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
from sklearn.metrics import mean_squared_error
XGB = XGBRegressor(n_estimators = 200)
XGB.fit(x_train, y_train)
pred_train = XGB.predict(x_train)
pred_test = XGB.predict(x_test)
print(mean_squared_error(y_train, pred_train))
print(mean_squared_error(y_test, pred_test))
```

```python
plt.scatter(y_test, pred_test)
plt.show()
```

```python
dummy_all_test = pd.DataFrame()
for col in categorical_columns.columns: 
    dummy = pd.get_dummies(test[col], prefix = col)
    if dummy_all_test.empty: 
        dummy_all_test = dummy
    else: 
        dummy_all_test = dummy_all_test.merge(dummy, how = 'outer', left_index = True, right_index = True)
```

```python
train_dummies = list(dummy_all.columns)
test_dummies = list(dummy_all_test.columns)
missing_dummies = np.setdiff1d(train_dummies, test_dummies)
for dummy in missing_dummies: 
    dummy_all_test[dummy] = 0
dummy_all_test = dummy_all_test[list(dummy_all.columns)]
```

```python
new_test = dummy_all_test.merge(test[numerical_cols.drop('SalePrice')], how = 'outer', left_index = True, right_index = True)
real_test_x = new_test.drop([
    'Id', 'MSSubClass', 'OverallCond', 'BsmtFinSF2', 'LowQualFinSF', 'BsmtHalfBath', 
    'KitchenAbvGr', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 
    'MoSold', 'YrSold'], axis = 1)

real_test_y = XGB.predict(real_test_x)
```

```python
test['SalePrice'] = real_test_y
test.to_csv('Oliver.csv', columns = ['Id', 'SalePrice'], index = False)
#print("Lowest RMSE: ", (-XGB.best_score_)**(1/2.0))
```
